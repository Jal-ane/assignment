# Iris Dataset Analysis

## Prepare the dataset

Use `numpy` or to load `data/iris.data`. Remember to look at the file contents to understand the data format.

```python
import numpy as np
data = np.genfromtxt('data/iris.data', delimiter = ',')
```

Create a map that returns the attribute name given the attribute index.

- 0 -> sepal length
- 1 -> sepal width
- 2 -> petal length
- 3 -> petal width

```python
mapping = {0:'Sepal Length', 1:'Sepal Width', 2:'Petal Length', 3:'Petal Width'}
```

## Data analysis

Write a function that prints the mean and another that prints the standard deviation of each attribute.

```python
attribute_data = np.loadtxt('data/iris.attributes', delimiter = ',')
def calc_mean(file):
    mean = np.mean(file, axis = 0)
    for i in range(len(mapping)):
        print(f'Mean of {mapping[i]} is {mean[i]}')
def calc_std(file):
    mean = np.std(file, axis = 0)
    for i in range(len(mapping)):
        print(f'Standard Deviation of {mapping[i]} is {mean[i]}')
calc_mean(attribute_data)
calc_std(attribute_data)  
```

Use the functions you defined above to print the mean of the attributes for each class.

```python
class_data = np.genfromtxt('data/iris.classes', dtype = str, encoding = None)
class_names = np.unique(class_data)
for class_name in class_names:
    fltr = class_data == class_name
    fltr_data = attribute_data[fltr]
    print(f'Class {class_name}:')
    calc_mean(fltr_data)
```

Now, let's plot this data. Use `matplotlib` to create scatter plots for petal and sepal. Use different colors for each class.

```python
import matplotlib.pyplot as plt

colors = ["red","blue","black"]

plt.figure(figsize = (10,6))
#petal plot
plt.subplot(1,2,1)
for class_name, color in zip(class_names, colors):
    fltr = class_data == class_name 
    plt.scatter(attribute_data[fltr,0], attribute_data[fltr,1], label = class_name, color =color)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal plot')
plt.legend()
#sepal plot
plt.subplot(1,2,2)
for class_name, color in zip(class_names, colors):
    fltr = class_data == class_name 
    plt.scatter(attribute_data[fltr,2], attribute_data[fltr,3], label = class_name, color =color)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Plot')
plt.legend()
plt.show()
```
